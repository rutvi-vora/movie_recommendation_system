### calculate the weighted rating

WR = (v ÷ (v+m)) × R + (m ÷ (v+m)) × C

```python
m = movies["vote_count"].quantile(0.9)
m
```

```python
C = movies["vote_average"].mean()
C
```

```python
movies_filtered = movies.copy().loc[movies["vote_count"] >= 0]
movies_filtered
```

```python
def weighted_rating(df, m=m, C=C):
    v = df["vote_count"]
    R = df["vote_average"]
    wr = (v /  (v + m)) * R + (m / (v + m)) * C
    return wr
```

```python
movies_filtered["weighted_rating"] = movies_filtered.apply(weighted_rating, axis=1)
```

```python
movies_filtered
```

```python
movies_filtered.sort_values("weighted_rating", ascending=False)[["title", "weighted_rating"]].head(10).to_dict()
```

### Content Based filtering

```python
import pandas as pd
# movies = pd.read_csv("data/movies_small.csv", sep=";")
movies = pd.read_csv("data/movies.csv")
```

```python
movies
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
# this will convert the text to numbers which will return coefficients means the values of each word in the text to compare the similarity
```

```python
tfidf = TfidfVectorizer(stop_words="english")

movies["overview"] = movies["overview"].fillna("")
```

```python
tfidf_matrix = tfidf.fit_transform(movies["overview"])
```

```python
pd.DataFrame(tfidf_matrix.toarray(), columns=tfidf.get_feature_names_out())
```

```python
#for first row word toretto has 0.33333 value that determines the frequency of this word in that particular movie description
# and how rare its frequency is compared to other movies in the datset. This is called term frequency and inverse document frequency
```

#### Similarity matrix

```python
from sklearn.metrics.pairwise import linear_kernel
similarity_matrix = linear_kernel(tfidf_matrix,  tfidf_matrix) # to have pairwise comparison we are mentioning duplicates
```

```python
similarity_matrix
```

```python
similarity_matrix.shape
```

## Find the most similar movies to a certain movie

```python
movie_title = "John Carter"
```

```python
idx = movies.loc[movies["title"] == movie_title].index[0]
idx
```

```python
scores = list(enumerate(similarity_matrix[idx]))
scores
```

```python
scores = sorted(scores, key=lambda x: x[1], reverse=True)
```

```python
movie_indices = [tpl[0] for tpl in scores[1:4]]
movie_indices
```

```python
list(movies["title"].iloc[movie_indices])
```

```python
def similar_movies(movie_title, nr_movies):
    idx = movies.loc[movies["title"] == movie_title].index[0]
    scores = list(enumerate(similarity_matrix[idx]))
    scores = sorted(scores, key=lambda x: x[1], reverse=True)
    movie_indices = [tpl[0] for tpl in scores[1:nr_movies + 1]]
    similar_titles = list(movies["title"].iloc[movie_indices])
    return similar_titles
```

```python
similar_movies("Kung Fu Panda 3", 3)
```
